```python
def supprimer_espace(nom_colonne):
    liste = []
    mot_colonne = nom_colonne.split()
    size = int(len(mot_colonne))
    for i in range(size):
        if i<size -1:
            liste.append(mot_colonne[i]+'_')
        else:
            liste.append(mot_colonne[i])
            
    nouveau_nom_colonne = ''.join(liste)
    return nouveau_nom_colonne
            
    
```

```python
supprimer_espace("Total price for in-state students living on campus 2013-14")
```

```python
supprimer_espace("Applicants total")
```

```python
liste_nouveaux_colonnes = []
colonnes = data.columns
for colonne in colonnes:
    nouveau_colonne = supprimer_espace(colonne)
    liste_nouveaux_colonnes.append(nouveau_colonne)
```

# Question 3 : Classement des universités selon le nombre de candidatures, d’admission et d’inscrits

# Nombre d’universités selon le nombre de candidature

```python
plt.figure(figsize = (14,8))

sns.histplot(data['Applicants_total'], bins = 40)
plt.title("Nombre d’universités selon le nombre de candidatures")
plt.xlabel("Nombre de candidatures")
plt.ylabel("Nombre des universités")
plt.axis([0,75000, 0,500])

Moyenne = "La moyenne est: {:.2f}".format(data['Applicants_total'].mean())
Mediane = "La mediane est: {:.2f}".format(data['Applicants_total'].median())

print(Moyenne + "\n")
print(Mediane)
```

# Nombre des universités selon le nombre d’admission

```python
plt.figure(figsize = (14,8))

sns.histplot(data['Admissions_total'], bins = 40)
plt.title("Nombre d’universités selon le nombre de candidatures")
plt.xlabel("Nombre de candidatures")
plt.ylabel("Nombre des universités")
plt.axis([0,75000, 0,500])

Moyenne = "La moyenne est: {:.2f}".format(data['Admissions_total'].mean())
Mediane = "La mediane est: {:.2f}".format(data['Admissions_total'].median())

print(Moyenne + "\n")
print(Mediane)
```

# Le nombre des universités selon le total des inscrits 

```python
plt.figure(figsize = (14,8))

sns.histplot(data['Enrolled_total'], bins = 30)
plt.title("Nombre d’universités selon le nombre de candidatures")
plt.xlabel("Nombre de candidatures")
plt.ylabel("Nombre des universités")
plt.axis([0,12000, 0,500])

Moyenne = "La moyenne est: {:.2f}".format(data['Enrolled_total'].mean())
Mediane = "La mediane est: {:.2f}".format(data['Enrolled_total'].median())

print(Moyenne + "\n")
print(Mediane)
```
