```python
import numpy as np
```

```python
arr = np.arange(1, 11)
arr
```

# **Fonctions universelles NumPy `Ufuncs`**

Les `Ufuncs` de NumPy rendent plus efficaces les opérations sur les tableaux.

Les `Ufuncs` de trois opérations majeures :
* Opérations arithmétiques
* Opérations d’agrégation
* Opérations de comparaison

# **Opérations arithmétiques**
* `add()`
* `substract()`
* `multiply()`
* `divide()`
* `floor_divide()`
* `power()`
* `mod()`

```python
np.add(arr, 5) # Ajouter 5 éléments
```

```python
np.subtract(arr, 2) # Enlève en 2
```

```python
np.multiply(arr, 10)
```

```python
np.divide(arr, 100)
```

```python
np.floor_divide(arr, 4)
```

```python
np.power(arr, 2)
```

```python
np.mod(arr, 2)
```

# **Opérations d’agrégation**
* `moyenne()`
* `médiane()`
* `std()`
* `var()`
* `min()`
* `max()`

```python
arr1 = np.array([1, 2, 3, 4, 5])
arr1
```

```python
np.mean(arr1)
```

```python
np.median(arr1)
```

```python
np.std(arr1)
```

```python
np.var(arr1)
```

```python
np.min(arr1)
```

```python
np.max(arr1)
```

# **Opérations de comparaison**
* `equal()`
* `less()`
* `less_equal()`
* `greater()`
* `greater_equal()`
* `not_equal()`

```python
a1 = np.array([1, 2, 3, 4])
a2 = np.array([1, 0, 3, 7])
a1, a2
```

```python
np.equal(a1, a2)
```

```python
np.less(a1, a2)
```

```python
np.less_equal(a1, a2)
```

```python
np.greater(a1, a2)
```

```python
np.greater_equal(a1, a2)
```

```python
np.not_equal(a1, a2)
```
