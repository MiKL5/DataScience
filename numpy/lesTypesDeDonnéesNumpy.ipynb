```python
import numpy as np
np.__version__
```

# **INTEGER**

```python
arr = np.array([1,2,3,4,5])
arr.dtype
```

Reduce memory usage

```python
arr1 = np.array([1, 2, 3, 4], dtype='int16')
arr1.dtype
```

```python
arr2 = np.array([1, 2, 3, 4], dtype='int8')
arr2.dtype
```

# **STRING**

```python
arr2 = np.array(['a', 'b', 'c', 'd'])
arr2.dtype
```

☝️ UNICODE LENGTH 1

```python
arr3 = np.array(['pomme', 'banane', 'cerise', 'dragon', 'cette liste de course s\'autodétruira dans 30 secondes'])
arr3.dtype
```
